```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import pandas as pd

# Now plot the graph

# create 3d figure
def plot_graph(X, Y, Z):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection = '3d')

    # Plot the line
    ax.plot(X, Y, Z)

    # Add labels and title
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('3D Line Plot')

    # Show the plot
    plt.show()
```
